Por medio de este trabajo se podrá visualizar la cantidad de mujeres y hombres, segun su **Autoidentificacion Etnica**(indigena, afroecuatoriano, montubio, mestizo, blanco u otro). Ademas de presentar tambien un porcentaje de hombres y mujeres bajo el mismo parametro y un total de personas. Tambien cuenta con un parametro que nos otorga una constante llamada indice de feminidad, la cual esta basada en la cantidad de por etnia sobre los hombres.

```python
import pandas as pd
```

```python
tabla1 = pd.read_csv('./data/POBLACION_SEGUN_AUTOIDENTIFICACION_ETNICA.csv')
```

```python
from bokeh.io import output_notebook, show
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource

output_notebook()
```

```python
group1 = tabla1.groupby('ETNIA')

source = ColumnDataSource(group1)

source.data
```

```python
from bokeh.layouts import gridplot

group1 = tabla1.groupby('ETNIA')

source = ColumnDataSource(group1)

options = dict(plot_width=800, plot_height=450, x_range=group1, tools="pan,wheel_zoom,box_zoom") 

graph1 = figure(title='POBLACION SEGUN AUTOIDENTIFICACION ETNICA', **options )

graph1.line(x='ETNIA', y='HOMBRES_NUMERO_mean', color="#FF7F50", legend='HOMBRES', source=source)
graph1.circle(x='ETNIA', y='HOMBRES_NUMERO_mean', color='#FF7F50', size=5.5, source=source)
graph1.line(x='ETNIA', y='MUJERES_NUMERO_mean', color="#9932CC", legend='MUJERES', source=source)
graph1.circle(x='ETNIA', y='MUJERES_NUMERO_mean', color="#9932CC", size=5.5, source=source)
graph = gridplot([[graph1]])

show(graph)
```

```python
from bokeh.layouts import gridplot

group1 = tabla1.groupby('ETNIA')

source = ColumnDataSource(group1)

options = dict(plot_width=800, plot_height=450, x_range=group1, tools="pan,wheel_zoom,box_zoom") 

graph1 = figure(title='POBLACION SEGUN AUTOIDENTIFICACION ETNICA', **options )

graph1.line(x='ETNIA', y='HOMBRES_%_mean', color="#7CFC00", legend='% HOMBRES', source=source)
graph1.circle(x='ETNIA', y='HOMBRES_%_mean', color='#7CFC00', size=5.5, source=source)
graph1.line(x='ETNIA', y='MUJERES_%_mean', color="#20B2AA", legend='% MUJERES', source=source)
graph1.circle(x='ETNIA', y='MUJERES_%_mean', color="#20B2AA", size=5.5, source=source)
graph = gridplot([[graph1]])

show(graph)
```

```python
group1 = tabla1.groupby('ETNIA')

source = ColumnDataSource(group1)

gph = figure(plot_height=250, plot_width=750, x_range=group1, title='POBLACION SEGUN AUTOIDENTIFICACION' )

gph.vbar(x='ETNIA', top='INDICE_DE_FEMINIDAD_mean', width=0.5, line_color="white", source=source)

gph.xgrid.grid_line_color = None
gph.xaxis.axis_label = "ETNIA - INDICE DE FEMINIDAD"
gph.y_range.start = 0

show(gph)
```
