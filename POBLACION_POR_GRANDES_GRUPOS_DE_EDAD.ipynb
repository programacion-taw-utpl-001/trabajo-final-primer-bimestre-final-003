En el conjunto de tablas y gráficas, se podrá observar **POBLACION POR GRANDES GRUPOS DE EDAD**, estos grupos de edad son: 0-14 años, 15-64 años, 65 años en adelante. Los datos observables corresponden a los años de censo de 1990, 2001 y 2010. Los datos están divididos entre hombres y mujeres con sus respectivos porcentajes, así como un total de cada censo.

```python
import pandas as pd
```

```python
pd.read_csv('./data/POBLACION_POR_GRANDES_GRUPOS_DE_EDAD.csv', encoding='latin-1') # se agrega el encoding='latin-1' para que se permita agregar datos con la ñ
```

```python
tabla = pd.read_csv('./data/POBLACION_POR_GRANDES_GRUPOS_DE_EDAD.csv', encoding='latin-1')
```

```python
from bokeh.io import output_notebook, show
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource

output_notebook()
```

```python
# Creamos una nueva tabla solo con los valores por año
tabla2 = tabla[tabla.AÑO_CENSAL == '1990']
tabla3 = tabla[tabla.AÑO_CENSAL == '2001']
tabla4 = tabla[tabla.AÑO_CENSAL == '2010']

# Unimos las tablas anteriores
tabla1 = pd.concat([tabla2, tabla3, tabla4])

# Agrupamos los valores en base a la columna "AÑO_CENSAL" 
group1 = tabla1.groupby('AÑO_CENSAL')

# Nos permite obtener Mapas de nombres columnas, secuencias, matrices.
source = ColumnDataSource(group1)

p = figure(plot_height=250, plot_width=750, x_range=group1,  title='AUMENTO POBLACIONAL - MUJERES')

p.vbar(x='AÑO_CENSAL', top='MUJERES_NUMERO_mean', width = 0.25, color='#FF69B4', source=source)

show(p)
```

```python
p = figure(plot_height=250, plot_width=750, x_range=group1,  title='AUMENTO POBLACIONAL - HOMBRES')

p.vbar(x='AÑO_CENSAL', top='HOMBRES_NUMERO_mean', width = 0.25, color='#FF6924', source=source)

show(p)
```

```python
from bokeh.layouts import gridplot

# Agrupamos los valores en base a la columna "AÑO_CENSAL" 
group1 = tabla.groupby('AÑO_CENSAL')

# Nos permite obtener Mapas de nombres columnas, secuencias, matrices.
source = ColumnDataSource(group1)

p = figure(plot_height=500, plot_width=500, y_range=group1,  title='AUMENTO POBLACIONAL - HOMBRES')
p2 = figure(plot_height=500, plot_width=500, y_range=group1,  title='AUMENTO POBLACIONAL - MUJERES')


p.hbar(y='AÑO_CENSAL', height=0.5, left=0,right='HOMBRES_NUMERO_mean', color='#FF6589', source=source)
p2.hbar(y='AÑO_CENSAL', height=0.5, left=0,right='MUJERES_NUMERO_mean', color='#FF6574', source=source)

graph = gridplot([[p, p2]])
show(graph)
```
