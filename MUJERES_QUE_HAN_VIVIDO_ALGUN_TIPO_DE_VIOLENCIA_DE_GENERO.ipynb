```python
tabla1 = pd.read_csv('./data/MUJERES_QUE_HAN_VIVIDO_ALGUN_TIPO_DE_VIOLENCIA_DE_GENERO.csv')
```

```python
from bokeh.io import output_notebook, show
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource


output_notebook()
```

```python
group1 = tabla1.groupby('AREA')

source = ColumnDataSource(group1)

source.data
```

```python
from bokeh.layouts import gridplot

group1 = tabla1.groupby('AREA')

source = ColumnDataSource(group1)

options = dict(plot_width=1000, plot_height=450, x_range=group1, tools="pan,wheel_zoom,box_zoom") 

graph1 = figure(title='MUJERES QUE HAN VIVIDO ALGUN TIPO DE VIOLENCIA DE GENERO', **options )

graph1.line(x='AREA', y='MUJERES_VIOLENTADAS_mean', color="#FFD700", legend='MUJERES VIOLENTADAS', source=source)
graph1.circle(x='AREA', y='MUJERES_VIOLENTADAS_mean', color='#FFD700', size=5.5, source=source)
graph1.line(x='AREA', y='TOTAL_MUJERES_INVESTIGADAS_mean', color="#FF4500", legend='CASOS INVESTIGADOS', source=source)
graph1.circle(x='AREA', y='TOTAL_MUJERES_INVESTIGADAS_mean', color="#FF4500", size=5.5, source=source)
graph = gridplot([[graph1]])

show(graph)
```

```python
group1 = tabla1.groupby('AREA')

source = ColumnDataSource(group1)

p = figure(plot_height=250, plot_width=750, x_range=group1,  title='ANALFABETISMO POR AREA Y GRUPOS DE EDAD', )

p.square(x='AREA', y='%_DE MUJERES_VIOLENTADAS_mean', color='#FF69B4', size=10, source=source)


p.xaxis.axis_label = "AREA - % DE MUJERES VIOLENTADAS"
p.y_range.start = 0

show(p)
```
