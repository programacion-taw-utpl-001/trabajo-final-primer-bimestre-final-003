```python
tabla1 = pd.read_csv('./data/INGRESOS_PROMEDIOS_DE_LOS_HOGARES_SEGUN_JEFATURA_DEL_HOGAR.csv')
```

```python
from bokeh.io import output_notebook, show
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource

output_notebook()
```

```python
group1 = tabla1.groupby('AREA')

source = ColumnDataSource(group1)

source.data
```

```python
from bokeh.layouts import gridplot

group1 = tabla1.groupby('AREA')

source = ColumnDataSource(group1)

options = dict(plot_width=450, plot_height=450, # Tamaño
                x_range=group1, # Mandamos el rango de x
               tools="pan,wheel_zoom,box_zoom") # Herramientas de bokeh a mostrar

graph1 = figure(title='JEFATURA FEMENINA', **options)
graph2 = figure(title='JEFATURA MASCULINA', **options)
graph3 = figure(title='TOTAL NACIONAL', **options)
graph4 = figure(title='DESIGUALDAD', **options)

graph1.line(x='AREA', y='INGRESO_PROMEDIO_DEL HOGAR_JEFATURA FEMENINA_mean', color="#00FF39", source=source)
graph1.circle(x='AREA', y='INGRESO_PROMEDIO_DEL HOGAR_JEFATURA FEMENINA_mean', color='#00008B', size=5.5, source=source)
graph2.line(x='AREA', y='INGRESO_PROMEDIO_DEL_HOGAR_JEFATURA_MASCULINA_mean', color="#000080", source=source)
graph2.circle(x='AREA', y='INGRESO_PROMEDIO_DEL_HOGAR_JEFATURA_MASCULINA_mean', color='#8B0000', size=5.5, source=source)
graph3.vbar(x='AREA', top='NACIONAL_mean', width=0.5, line_color="white", source=source)
graph4.vbar(x='AREA', top='DESIGUALDAD_mean', width=0.5, line_color="white", color="green", source=source)
graph = gridplot([[graph1, graph2, graph3, graph4]])

show(graph)
```
