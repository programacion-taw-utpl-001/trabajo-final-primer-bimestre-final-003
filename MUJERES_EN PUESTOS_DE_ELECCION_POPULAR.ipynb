```python
tabla1 = pd.read_csv('./data/MUJERES_EN PUESTOS_DE_ELECCION_POPULAR.csv')
```

```python
from bokeh.io import output_notebook, show
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource

output_notebook()
```

```python
group1 = tabla1.groupby('DIGNIDADES')

# Nos permite obtener Mapas nombres de columnas a secuencias o matrices.
source = ColumnDataSource(group1)

# Mostramos los nombres de los datos albergados en source - donde buscamos los nombre de los datos que buscamos en concreto
source.data
```

```python
from bokeh.layouts import gridplot

group1 = tabla1.groupby('DIGNIDADES')

source = ColumnDataSource(group1)

options = dict(plot_width=1500, plot_height=450, x_range=group1, tools="pan,wheel_zoom,box_zoom") 

graph1 = figure(title='CANDIDATURAS MUJERES(VERDE) - CANDIDATURAS HOMBRES(ROJO)', **options )

graph1.line(x='DIGNIDADES', y='CANDIDATURAS_HOMBRES_NUMERO_mean', color="#FF0000", source=source)
graph1.line(x='DIGNIDADES', y='CANDIDATURAS_MUJERES_NUMERO_mean', color="#ADFF2F", source=source)

graph = gridplot([[graph1]])

show(graph)
```

```python
from bokeh.layouts import gridplot

group1 = tabla1.groupby('DIGNIDADES')

source = ColumnDataSource(group1)

options = dict(plot_width=1500, plot_height=450, x_range=group1, tools="pan,wheel_zoom,box_zoom") 

graph1 = figure(title='% CANDIDATURAS HOMBRES', **options )
graph2 = figure(title='% CANDIDATURAS MUJERES', **options )

graph1.vbar(x='DIGNIDADES', top='CANDIDATURAS_HOMBRES_NUMERO_mean', width=0.5, line_color="white", color="#FF0000", source=source)
graph2.vbar(x='DIGNIDADES', top='CANDIDATURAS_MUJERES_NUMERO_mean', width=0.5, line_color="white", color="#ADFF2F", source=source)

graph = gridplot([[graph1, graph2]])

show(graph)
```
